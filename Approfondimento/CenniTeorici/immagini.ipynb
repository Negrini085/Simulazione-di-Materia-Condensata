```python
import numpy as np
import matplotlib.pyplot as plt


L = 10

angles = np.zeros(shape=(10, 10))
for i in range(L):
    for j in range(L):
        angles[i][j] = np.pi/2


x, y = np.meshgrid(np.arange(1, L+1), np.arange(1, L+1))

u = np.cos(angles)
v = np.sin(angles)

plt.figure(figsize=(6,6))
plt.quiver(x, y, u, v, angles='xy', scale_units='xy', scale=1.5, color='b')
plt.xlim(0, L+1)
plt.ylim(0, L+1)
plt.xticks(range(1,L+1))
plt.yticks(range(1,L+1))
plt.grid(True, linestyle='--', alpha=0.5, color='gray')
plt.title("Modello XY: ground state", fontsize=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


L = 10

angles = np.zeros(shape=(10, 10))
for i in range(L):
    for j in range(L):
        angles[i][j] = (j+1)/L*np.pi/6


x, y = np.meshgrid(np.arange(1, L+1), np.arange(1, L+1))

u = np.cos(angles)
v = np.sin(angles)

plt.figure(figsize=(6,6))
plt.quiver(x, y, u, v, angles='xy', scale_units='xy', scale=1.5, color='b')
plt.xlim(0, L+1)
plt.ylim(0, L+1)
plt.xticks(range(1,L+1))
plt.yticks(range(1,L+1))
plt.grid(True, linestyle='--', alpha=0.5, color='gray')
plt.title("Modello XY: spin-wave", fontsize=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensioni del reticolo
L = 10

# Crea le coordinate del reticolo centrato
x, y = np.meshgrid(np.linspace(-L, L, 2*L), np.linspace(-L, L, 2*L))

# Calcola gli angoli per il vortice con winding number = 1
angles = np.arctan2(y, x)

# Componenti dei vettori (freccette), normalizzate per chiudersi su se stesse
u = -np.sin(angles)  # Componente x
v = np.cos(angles)   # Componente y

# Disegna il reticolo con le frecce
plt.figure(figsize=(6,6))
plt.quiver(x, y, u, v, angles='xy', scale_units='xy', scale=1.2, color='b')
plt.xlim(-L, L)
plt.ylim(-L, L)
plt.xticks([])
plt.yticks([])
plt.grid(True, linestyle='--', alpha=0.5, color='gray')
plt.title("Modello XY: esempio di vortice", fontsize = 20)
plt.show()
```
