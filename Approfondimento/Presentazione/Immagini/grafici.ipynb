### Magnetizzazione

```python
T = np.linspace(0.3, 2.2, 200)
beta = 1/T
h = 0.0
J = 1.0

m_0 = np.sinh(beta * h)/np.sqrt(np.exp(-4*beta*J) + np.sinh(beta * h)**2)


fig = plt.figure(figsize = (15, 8))

plt.plot(T, m_0, label = 'Soluzione esatta')
plt.errorbar(t_camp, magn_1000, yerr = magn_1000_err, linestyle = 'None', fmt = 'o', markerfacecolor='none', label = r'$N_{s}\,=\,1000$')
plt.errorbar(t_camp, magn_3000, yerr = magn_3000_err, linestyle = 'None', fmt = 'x', label = r'$N_{s}\,=\,3000$')
plt.errorbar(t_camp, magn_6000, yerr = magn_6000_err, linestyle = 'None', fmt = '^', markerfacecolor='none', label = r'$N_{s}\,=\,6000$')
plt.errorbar(t_camp, magn_10000, yerr = magn_10000_err, linestyle = 'None', fmt = 'D', markerfacecolor='none', label = r'$N_{s}\,=\,10000$')

plt.title("Magnetizzazione Ising1D: h = 0.0", fontsize = 20)
plt.xlabel("Temperatura", fontsize = 15)
plt.ylabel("m", fontsize = 15)

plt.legend(loc = 'best', fontsize = 14)
plt.grid(True)
plt.show()
```

### Energia interna

```python
T = np.linspace(0.2, 2.0, 200)
beta = 1/T
h = 0.02
J = 1.0

rad = np.sqrt(np.exp(-2 * beta * J) + np.exp(2 * beta * J) * pow(np.sinh(beta * h), 2))

l1 = np.exp(beta * J) * np.cosh(beta * h) + rad
term1 = np.exp(beta * J) * (J * np.cosh(beta * h) + h * np.sinh(beta * h))
term2 = -2 * J * np.exp(-2 * beta * J) + 2 * J * np.exp(2 * beta * J) * pow(np.sinh(beta * h), 2) + 2 * h * np.exp(2 * beta * J) * np.sinh(beta * h) * np.cosh(beta * h)

ene = - (1/l1) * (term1 + term2/(2 * rad))


fig = plt.figure(figsize = (15, 8))

plt.plot(T, ene, label = 'Soluzione esatta')

plt.errorbar(t_camp, ene_m_1000, yerr = ene_m_1000_err, linestyle = 'None', fmt = 'o', markerfacecolor='none', label = r'$N_{s}\,=\,1000$')
plt.errorbar(t_camp, ene_m_3000, yerr = ene_m_3000_err, linestyle = 'None', fmt = 'x', markerfacecolor='none', label = r'$N_{s}\,=\,3000$')
plt.errorbar(t_camp, ene_m_6000, yerr = ene_m_6000_err, linestyle = 'None', fmt = '^', markerfacecolor='none', label = r'$N_{s}\,=\,6000$')
plt.errorbar(t_camp, ene_m_10000, yerr = ene_m_10000_err, linestyle = 'None', fmt = 'D', markerfacecolor='none', label = r'$N_{s}\,=\,10000$')


plt.title("Energia interna per spin Ising1D: h = 0.02", fontsize = 20)
plt.xlabel("Temperatura", fontsize = 15)
plt.ylabel(r"$U\left(T\right)$", fontsize = 15)

plt.legend(loc = 'best', fontsize = 14)
plt.grid(True)
plt.show()
```

### Suscettivit magnetica

```python
T = np.linspace(0.2, 2.0, 200)
beta = 1/T
h = 0.02
J = 1.0


num = np.cosh(beta * h) * np.exp(-4 * beta * J)
den = pow(np.exp(-4 * beta * J) + pow(np.sinh(beta * h), 2), 1.5)
chi = beta  * num/den


fig = plt.figure(figsize = (15, 8))

plt.plot(T, chi, label = 'Soluzione esatta')

plt.errorbar(t_camp, chi_m_1000, yerr = chi_m_1000_err, linestyle = 'None', fmt = 'o', markerfacecolor='none', label = r'$N_{s}\,=\,1000$')
plt.errorbar(t_camp, chi_m_3000, yerr = chi_m_3000_err, linestyle = 'None', fmt = 'o', markerfacecolor='none', label = r'$N_{s}\,=\,3000$')
plt.errorbar(t_camp, chi_m_6000, yerr = chi_m_6000_err, linestyle = 'None', fmt = 'o', markerfacecolor='none', label = r'$N_{s}\,=\,6000$')
plt.errorbar(t_camp, chi_m_10000, yerr = chi_m_10000_err, linestyle = 'None', fmt = 'o', markerfacecolor='none', label = r'$N_{s}\,=\,10000$')

plt.title("Suscettivit Ising1D: h = 0.02", fontsize = 20)
plt.xlabel("Temperatura", fontsize = 15)
plt.ylabel(r"$\chi\left(T\right)$", fontsize = 15)

plt.legend(loc = 'best', fontsize = 14)
plt.grid(True)
plt.show()


```
