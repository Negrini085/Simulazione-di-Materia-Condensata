### Calore specifico

```python
T = np.linspace(0.3, 2.2, 200)
beta = 1/T
h = 0.02
J = 1.0

beta_h = beta * h
beta_J = beta * J

rad = np.sqrt(np.exp(-2 * beta_J) + np.exp(2 * beta_J) * pow(np.sinh(beta_h), 2))

l1 = np.exp(beta_J) * np.cosh(beta_h) + rad

# Derivate prime
der1_t1 = np.exp(beta_J) * (J * np.cosh(beta_h) + h * np.sinh(beta_h))
der_arg_rad = -2 * J * np.exp(-2 * beta_J) + 2 * np.exp(2 * beta_J) * np.sinh(beta_h) * (J * np.sinh(beta_h) + h * np.cosh(beta_h))
der1_t2 = (1/(2 * rad)) * der_arg_rad
der1 = der1_t1 + der1_t2


# Derivate seconde
der2_t1 = np.exp(beta_J) * (np.cosh(beta_h) * (J**2 + h**2) + 2 * h * J * np.sinh(beta_h))
der2_arg_rad_t1 = 4 * J**2 * np.exp(-2 * beta_J)
der2_arg_rad_t2 = 2 * J * np.exp(2 * beta_J) * (2 * J * pow(np.sinh(beta * h), 2) + h * np.sinh(2 * beta_h))
der2_arg_rad_t3 = np.exp(2 * beta_J) * (2 * J * h * np.sinh(2 * beta * h) + 2 * h**2 * np.cosh(2 * beta_h))
der2_arg_rad = der2_arg_rad_t1 + der2_arg_rad_t2 + der2_arg_rad_t3
der2_t2 = - (1/(4 * pow(rad, 3))) * pow(der_arg_rad, 2) + (1/(2 * rad)) * der2_arg_rad
der2 = der2_t1 + der2_t2


cv = pow(beta, 2)/l1 * (-1/l1 * der1**2 + der2)


fig = plt.figure(figsize = (15, 8))

plt.plot(T, cv, label = 'Soluzione esatta')

plt.errorbar(t_camp, cp_m_1000, yerr = cp_m_1000_err, linestyle = 'None', fmt = 'o', markerfacecolor='none', label = r'$N_{s}\,=\,1000$')
plt.errorbar(t_camp, cp_m_3000, yerr = cp_m_3000_err, linestyle = 'None', fmt = 'x', markerfacecolor='none', label = r'$N_{s}\,=\,3000$')
plt.errorbar(t_camp, cp_m_6000, yerr = cp_m_6000_err, linestyle = 'None', fmt = '^', markerfacecolor='none', label = r'$N_{s}\,=\,6000$')
plt.errorbar(t_camp, cp_m_10000, yerr = cp_m_10000_err, linestyle = 'None', fmt = 'D', markerfacecolor='none', label = r'$N_{s}\,=\,10000$')

plt.title("Calore specifico per spin Ising1D: h = 0.02", fontsize = 20)
plt.xlabel("Temperatura", fontsize = 15)
plt.ylabel(r"$c\left(T\right)$", fontsize = 15)

plt.legend(loc = 'best', fontsize = 14)
plt.grid(True)
plt.show()
```

### Magnetizzazione

```python
T = np.linspace(0.8, 3.7, 500)
J = 1  
Tc = 2 * J/math.log(1 + np.sqrt(2))

# Funzione per calcolare la magnetizzazione (soluzione di Onsager)
def magnOnsager(T, J, Tc):

    if T > Tc:
        m = 0
    else:
        m = np.sinh(2*J/T)
        m = (1 - m**(-4))**(1/8)
    
    return m


# Calcolo della magnetizzazione per ogni temperatura
m = []
for t in T:
    m.append(magnOnsager(t, J, Tc))

# Creazione del grafico
plt.figure(figsize=(15, 8))

plt.plot(T, m, label=r'Soluzione esatta', color='b')

plt.errorbar(t_camp, magn_100, yerr = magn_100_err, linestyle = 'None', fmt = 'o', markerfacecolor='none', label = r'$100 \times 100$')
plt.errorbar(t_camp, magn_200, yerr = magn_200_err, linestyle = 'None', fmt = 'x', markerfacecolor='none', label = r'$200 \times 200$')
plt.errorbar(t_camp, magn_300, yerr = magn_300_err, linestyle = 'None', fmt = '^', markerfacecolor='none', label = r'$300 \times 300$')
plt.errorbar(t_camp, magn_400, yerr = magn_400_err, linestyle = 'None', fmt = 'D', markerfacecolor='none', label = r'$400 \times 400$')
plt.errorbar(t_camp, magn_500, yerr = magn_500_err, linestyle = 'None', fmt = 's', markerfacecolor='none', label = r'$500 \times 500$')

plt.axvline(x=Tc, color='r', linestyle='--', label=f'Tc = {Tc:.3f}')
plt.xlabel('Temperatura', fontsize = 20)
plt.ylabel('m(T)', fontsize = 20)
plt.title('Magnetizzazione Ising 2D', fontsize = 25)
plt.tick_params(axis='both', which='major', labelsize=15)

plt.legend(loc = 'lower left', fontsize = 15)
plt.grid(True)
plt.show()
```

### Energia

```python
T = np.linspace(0.8, 3.7, 500)
J = 1  
Tc = 2 * J/math.log(1 + np.sqrt(2))

# Funzione per calcolare la magnetizzazione (soluzione di Onsager)
def eneOnsager(T, J):
    
    # Fattore che compare in tutti gli argomenti
    betaJ = J/T
    
    # Componenti per calcolo energia
    k = 2*np.sinh(2*betaJ)/pow(np.cosh(2*betaJ), 2)
    k1 = 2*pow(np.tanh(2*betaJ), 2) - 1
    intE = ellipk(k**2)

    ene = -J * np.cosh(2*betaJ)/np.sinh(2*betaJ) * (1 + 2/np.pi * k1 * intE)

    return ene


# Calcolo della magnetizzazione per ogni temperatura
en = []
for t in T:
    en.append(eneOnsager(t, J))

# Creazione del grafico
plt.figure(figsize=(15, 8))

plt.plot(T, en, label=r'Soluzione esatta', color='b')

plt.errorbar(t_camp, ene_100, yerr = ene_100_err, linestyle = 'None', fmt = 'o', markerfacecolor='none', label = r'$100 \times 100$')
plt.errorbar(t_camp, ene_200, yerr = ene_200_err, linestyle = 'None', fmt = 'x', markerfacecolor='none', label = r'$200 \times 200$')
plt.errorbar(t_camp, ene_300, yerr = ene_300_err, linestyle = 'None', fmt = '^', markerfacecolor='none', label = r'$300 \times 300$')
plt.errorbar(t_camp, ene_400, yerr = ene_400_err, linestyle = 'None', fmt = 'D', markerfacecolor='none', label = r'$400 \times 400$')
plt.errorbar(t_camp, ene_500, yerr = ene_500_err, linestyle = 'None', fmt = 's', markerfacecolor='none', label = r'$500 \times 500$')

plt.axvline(x=Tc, color='r', linestyle='--', label=f'Tc = {Tc:.3f}')
plt.xlabel('Temperatura', fontsize = 20)
plt.ylabel('U(T)', fontsize = 20)
plt.title('Energia Ising 2D: valor vero', fontsize = 25)
plt.tick_params(axis='both', which='major', labelsize=15)

plt.legend(loc = 'upper left', fontsize = 15)
plt.grid(True)
plt.show()
```

```python
# Funzione per calcolare acceptance rate Metropolis
def accR(T, deltE):
    
    # Fattore che compare in tutti gli argomenti
    beta = 1/T

    return np.exp(-beta * deltE)



# Creazione del grafico
plt.figure(figsize=(15, 8))
deltaE = np.linspace(0.0, 8.0, 500)

# T = 1.0
plt.plot(-deltaE, np.zeros(len(deltaE))+1, color='blue')
plt.plot(deltaE, accR(1.0, deltaE), color='blue', label = r'$\beta\,=\,1.0$')

# T = 2.0
plt.plot(-deltaE, np.zeros(len(deltaE))+1, color='red')
plt.plot(deltaE, accR(2.0, deltaE), color='red', label = r'$\beta\,=\,0.5$')

# T = 4.0
plt.plot(-deltaE, np.zeros(len(deltaE))+1, color='green')
plt.plot(deltaE, accR(4.0, deltaE), color='green', label = r'$\beta\,=\,0.25$')

# T = 8.0
plt.plot(-deltaE, np.zeros(len(deltaE))+1, color='orange')
plt.plot(deltaE, accR(8.0, deltaE), color='orange', label = r'$\beta\,=\,0.125$')


plt.xlabel(r'$\Delta E$', fontsize = 20)
plt.ylabel('AR', fontsize = 20)
plt.title('AR Metropolis', fontsize = 25)
plt.tick_params(axis='both', which='major', labelsize=15)

plt.legend(loc = 'upper right', fontsize = 15)
plt.grid(True)
plt.show()
```
