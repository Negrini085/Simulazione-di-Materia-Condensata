### T = 0.5

```python
L = 15
angles = np.loadtxt("conf/conf_t0.5_size100.out")
angles = angles[20:35, 20:35]



x, y = np.meshgrid(np.arange(1, L+1), np.arange(1, L+1))

u = np.cos(angles)
v = np.sin(angles)

plt.figure(figsize=(6,6))
plt.quiver(x, y, u, v, angles='xy', scale_units='xy', scale=1.5, color='b')
plt.xlim(0, L+1)
plt.ylim(0, L+1)
plt.xticks(range(1,L+1))
plt.yticks(range(1,L+1))
plt.grid(True, linestyle='--', alpha=0.5, color='gray')
plt.title("Modello XY: T = 0.5", fontsize=20)
plt.show()
```

```python
temp = np.array(['0.25','0.5','0.75','1.0','1.25','1.5','1.75','2.0','2.25','2.3','2.35','2.4'])
t_camp = np.array([0.25,0.5,0.75,1.0,1.25,1.5,1.75,2.0,2.25,2.3,2.35,2.4])
```

```python
size = '50'

ene_100 = np.zeros(len(temp))
ene_100_err = np.zeros(len(temp))


# Ciclo sulle varie temperature
for i in range(0, len(temp)):

    name = 'obs/obs_t' + temp[i] + '_size' + size + '.out'
    obs = np.loadtxt(name)


    # Valuto i valori medi
    ene_100[i] = np.mean(obs[:, 1])
    ene_100_err[i] = np.std(obs[:, 1], ddof=1)
```

```python
plt.plot(t_camp, ene_100)
```
